```python
#Import all necessary libraries
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import *
```

## 2-input AND gate

```python
#initialise Dataset for AND logic implementation
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [[0], [0], [0], [1]]
```

```python
# #Make a 2 layer model with input size =2 
# model = Sequential()
# model.add(Dense(16,input_dim=2,activation="relu"))
# model.add(Dense(1,activation="sigmoid"))
# model.summary()
```

```python
model = Sequential()
model.add(Dense(1,input_shape=(2,),activation='sigmoid'))
model.summary()
```

```python
#Use adam optimizer and b_c for loss function
model.compile(loss='binary_crossentropy', optimizer='SGD',metrics=['binary_accuracy'])
```

```python
model.fit(X, y, epochs=700,verbose=1)#Train the model
```

```python
#Check the loss and accuracy 
loss, accuracy = model.evaluate(X, y)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```

```python
#Test it out manually, the result with the highest probability is the output of AND gate
predictions = model.predict(X)
print(predictions)
```

# XOR gates
### We have to go with multilayer perceptron for XOR gate

### i. 2 Input XOR gate


```python
#dataset for 4 input OR gate
X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

y = [
    [0],
    [1],
    [1],
    [0]
]



#Make a 2 layer model with input size =2 
model = Sequential()
model.add(Dense(16,input_dim=2,activation="tanh"))
model.add(Dense(1,activation="sigmoid"))
model.summary()
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['binary_accuracy'])#using appropriate hyperparameters
model.fit(X, y, epochs=700,verbose=1)#Train the model
#Check the loss and accuracy 
loss, accuracy = model.evaluate(X, y)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
#Test it out manually, the results with the highest probability is the output of OR gate
predictions = model.predict(X)
print(predictions)
```

### 3 input XOR gate
