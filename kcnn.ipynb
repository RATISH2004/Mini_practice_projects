```python
import matplotlib.pyplot as plt
from sklearn import preprocessing
ht = [145,153,155,165,168,188,159]
wt = [70,65,63,58,60,56,90]
normalized_arrht  = preprocessing.normalize([ht])
normalized_wt = preprocessing.normalize([wt])
# print(type(normalized_arrht))
classes = [1,1,0,0,1,0,0]
plt.scatter(normalized_arrht,normalized_wt,c=classes)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib
data = list(zip(normalized_arrht,normalized_wt))
data = np.array(data).reshape(-1, 2)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(data,classes)
new_x =0.40
new_y= 0.40
newpoint = [(new_x,new_y)]
prediction = knn.predict(newpoint)
plt.scatter(np.append(normalized_arrht,new_x), np.append(normalized_wt,new_y),c=classes+[prediction[0]])
plt.show()


```
